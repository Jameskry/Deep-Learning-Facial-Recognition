```python
import dlib
import cv2
import tensorflow as tf
import matplotlib.image as mpimg
import scipy.misc
import glob
import numpy as np

# HOG face detector
face_detector = dlib.get_frontal_face_detector()

# Shape predictor
shape_predictor = dlib.shape_predictor('shape_predictor_68_face_landmarks.dat')

# 128 hidden node - model
face_recognition_model = dlib.face_recognition_model_v1('dlib_face_recognition_resnet_model_v1.dat')

images = []

for file_name in glob.glob('Khoi/*.jpg'):
    im = mpimg.imread(file_name)
    faces = face_detector(im, 1)
    shapes = [shape_predictor(im, face) for face in faces]
    face_ids = [face_recognition_model.compute_face_descriptor(im, shape, 1) for shape in shapes]
    face_ids_array = np.array(face_ids)
    face_ids_list = face_ids_array.tolist()[0]
    images.append(face_ids_list)

for file_name in glob.glob('KD/*.jpg'):
    im = mpimg.imread(file_name)
    faces = face_detector(im, 1)
    shapes = [shape_predictor(im, face) for face in faces]
    face_ids = [face_recognition_model.compute_face_descriptor(im, shape, 1) for shape in shapes]
    face_ids_array = np.array(face_ids)
    face_ids_list = face_ids_array.tolist()[0]
    images.append(face_ids_list)

labels = []
khoi = [1, 0]
kd = [0, 1]

for i in range(150):
    labels.append(khoi)

for i in range(150):
    labels.append(kd)

Inputs = np.array(images)
Labels = np.array(labels)

print (Inputs.shape)
print (Labels.shape)
```

```python
import csv

with open("Inputs.csv", "w", newline='') as f:
    writer = csv.writer(f)
    writer.writerows(Inputs)
    
with open("Labels.csv", "w", newline='') as f:
    writer = csv.writer(f)
    writer.writerows(Labels)
```

```python
# Hyperparameter
hidden_size1 = 200
hidden_size2 = 100


# Create placeholders
X  = tf.placeholder(tf.float32, [None, 128])
Y_ = tf.placeholder(tf.float32, [None, 2])
lr = tf.placeholder(tf.float32)
pkeep = tf.placeholder(tf.float32)

# Create variables
w1 = tf.Variable(tf.truncated_normal([128, hidden_size1], stddev=0.1), tf.float32)
b1 = tf.Variable(tf.ones([hidden_size1])/10)

w2 = tf.Variable(tf.truncated_normal([hidden_size1, hidden_size2], stddev=0.1), tf.float32)
b2 = tf.Variable(tf.ones([hidden_size2])/10)

w3 = tf.Variable(tf.truncated_normal([hidden_size2, 2], stddev=0.1), tf.float32)
b3 = tf.Variable(tf.ones([2])/10)

Y1 = tf.matmul(X, w1) + b1
Y1 = tf.nn.relu(Y1)

Y2 = tf.matmul(Y1, w2) + b2
Y2 = tf.nn.relu(Y2)

Ydrop = tf.nn.dropout(Y2, pkeep)

Ylogits = tf.matmul(Ydrop, w3) + b3
Y = tf.nn.softmax(Ylogits)

# Loss (Cross-entropy)
loss = tf.nn.softmax_cross_entropy_with_logits(logits=Ylogits, labels=Y_)

# Optimizer
optimizer = tf.train.AdamOptimizer(lr).minimize(loss)

# Accuracy
is_correct = tf.equal(tf.argmax(Y,1), tf.argmax(Y_,1))
accuracy = tf.reduce_mean(tf.cast(is_correct, tf.float32))

init = tf.global_variables_initializer()
saver = tf.train.Saver()
```

```python
with tf.Session() as sess:
    sess.run(init)
    
    train_dict = {X: Inputs, Y_: Labels, lr: 0.0001, pkeep: 0.9}
    # Training
    for i in range(1000):
        sess.run(optimizer, feed_dict=train_dict)
        
        # Accuracy
        if i % 100 == 0:
            A, C = sess.run([accuracy, loss], feed_dict=train_dict)
            print ("At iteration: %s, Loss: %s, Accuracy: %s" % (i,sum(C),A))
    
    save_path = saver.save(sess, "model.ckpt")
    print("Model saved in file: %s" % save_path)
```

```python
import tensorflow as tf
import numpy as np
import cv2

test_dict = {X: Inputs_test, Y_: Labels_test, lr: 0.0001, pkeep: 0.9}

with tf.Session() as sess:
    # Restore variables from disk.
    saver.restore(sess, "model.ckpt")
    print("Model restored.")
    
    Y_predict = sess.run(Y, feed_dict=test_dict)

    for i in range(len(Labels_test)):
        if np.argmax(Y_predict[i]) == 0:
            predicted_face = "Khoi face"
        elif np.argmax(Y_predict[i]) == 1:
            predicted_face = "KD face"


        if np.argmax(Labels_test[i]) == 0:
            actual_face = "Khoi face"
        elif np.argmax(Labels_test[i]) == 1:
            actual_face = "KD face"


        print ("Predicted: %s" % (predicted_face))
        print ("Actual: %s" % (actual_face))
        
        image = images_print[i]
        font = cv2.FONT_HERSHEY_SIMPLEX
        cv2.putText(image,predicted_face,(500,500), font, 4,(255,255,0),1,cv2.LINE_AA)
        cv2.imshow("Face", image)
        cv2.waitKey(0)            # Press Enter to see next image.

    cv2.destroyAllWindows()
```
